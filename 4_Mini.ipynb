# ü©∫ Minimundo 4 ‚Äî Rela√ß√£o entre IMC e Press√£o Arterial

## üìò Contexto

Uma cl√≠nica busca entender como a **idade** e o **√≠ndice de massa corporal (IMC)** se relacionam com a **press√£o arterial sist√≥lica** de seus pacientes.

---

## üß© Tarefas realizadas

* Calcular m√©dias e correla√ß√µes entre **IMC** e **press√£o arterial**
* Gerar **gr√°fico de dispers√£o** (*idade √ó press√£o*)
* Criar **mapa de calor** com a **matriz de correla√ß√£o**

---

## üß∞ Bibliotecas utilizadas

* `pandas`
* `seaborn`

---

## üìÇ Arquivo base

**pressao_arterial.csv**

| paciente | idade | imc  | pressao_sistolica |
| -------- | ----- | ---- | ----------------- |
| 1        | 25    | 22.4 | 118               |
| 2        | 40    | 27.1 | 132               |
| 3        | 55    | 30.2 | 146               |
| 4        | 65    | 31.8 | 155               |
| 5        | 35    | 25.7 | 128               |

---


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Lendo arquivo
df = pd.read_csv("dados/pressao_arterial.csv")

# Calculando M√©dia e correla√ß√£o entre IMC e Press√£o Arterial
Media_IMC = df['imc'].mean()
Media_Press = df['pressao_sistolica'].mean()
correlacao_imc = df['imc'].corr(df['pressao_sistolica'])
correlacao_idade = df['idade'].corr(df['pressao_sistolica'])

#Imprimindo Valores
print(f"M√©dia do IMC: {Media_IMC:.2f}")
print(f"M√©dia da press√£o arterial: {Media_Press:.2f}")
print(f"Correla√ß√£o entre IMC e Press√£o arterial: {correlacao_imc:.2f}")
print(f"Correla√ß√£o entre Idade e Press√£o arterial: {correlacao_idade:.2f}")

# Criando gr√°fico de dispers√£o
sns.regplot(data=df, x="idade", y="pressao_sistolica")
plt.title("Correla√ß√£o entre Idade e Press√£o Arterial")
plt.xlabel("Idade")
plt.ylabel("Press√£o Arterial Sist√≥lica (mmHg)")
plt.show()

# Criando matriz de correla√ß√£o
matriz_corr = df.corr()

# Plotando o heatmap
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(matriz_corr, fmt='.2f', square=True, linecolor='white', annot=True, cmap="coolwarm")
plt.title("Matriz de Correla√ß√£o - Todas as Vari√°veis")
plt.tight_layout()
plt.show()
```
