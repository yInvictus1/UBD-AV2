## üíâ Minimundo 6 ‚Äî Efic√°cia de Vacinas

**Contexto:**  
Um estudo avalia a **resposta imune** (n√≠vel de anticorpos) gerada por diferentes vacinas em pessoas de idades variadas.

**Tarefas realizadas:**
- Calcular m√©dia de anticorpos por tipo de vacina.  
- Plotar boxplot comparativo e mapa de calor.  
- Gerar gr√°fico de dispers√£o (*idade √ó anticorpos*).

**Bibliotecas utilizadas:**  
`pandas`, `seaborn`

**Arquivo base:** `vacinas.csv`

| vacina | idade | anticorpos |
|:-------|:------|:-----------|
| A | 30 | 1150 |
| A | 40 | 1280 |
| B | 30 | 1400 |
| B | 45 | 1350 |
| C | 35 | 1000 |

---

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("dados/vacinas.csv")

media = df.groupby("vacina")["anticorpos"].agg(["mean"])
print(f"media de anticorpos por vacina: \n{media}")

# Criar o boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="vacina", y="anticorpos", data=df) # x: vari√°vel categ√≥rica, y: vari√°vel num√©rica
plt.title("Compara√ß√£o dos N√≠veis de Anticorpos por Vacina")
plt.xlabel("Tipo de Vacina")
plt.ylabel("Total de Anticorpos")
plt.show()

# Criando matriz de correla√ß√£o
df_numeric = df[['idade', 'anticorpos']]
correlacao = df_numeric.corr()

# Criar heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, fmt='.2f')
plt.title('Matriz de Correla√ß√£o: Idade √ó Anticorpos')
plt.show()

#Criando grafico de dispers√£o
sns.regplot(data=df, x="anticorpos", y="idade")
plt.title("Correla√ß√£o entre Idade e Anticorpos")
plt.xlabel("Anticorpos")
plt.ylabel("Idade")
plt.show()
```
